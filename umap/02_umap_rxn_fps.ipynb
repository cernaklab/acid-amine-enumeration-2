### compute umap reduction of 1024-bit reaction fingerprints

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import pickle

import sys
import umap
import numpy as np
import ipywidgets
```

```python
amd33_raw = np.load("./acam33_rxn_fps_1024.npy")
```

```python
# %%time
all_n_n = [10,50,100,200,500,1000]

for n_n in all_n_n:
    
    reducer = umap.UMAP(n_neighbors=n_n,low_memory=True)
    embedding = reducer.fit_transform(amd33_raw)
    np.save(f"umap_acid_amine_33_nn_{n_n}.npy",embedding)
    reducer = []
    embedding = []
```

```python
# %%time
all_n_n = [10,50,100,200,500]

for n_n in all_n_n:
    
    # this will look different because umap runs differently for every iteration
    embedding = np.load(f"./umap_acid_amine_33_nn_{n_n}.npy")
    
    plt.figure(figsize=(10,8))
    plt.scatter(embedding[:,0],embedding[:,1])
    plt.title(str(n_n))
```
