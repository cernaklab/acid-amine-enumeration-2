```python

import pandas as pd
import matplotlib.pyplot as plt
from tqdm.contrib.concurrent import process_map
import numpy as np
import pickle

# from rdkit import DataStructs
import sys

import numpy as np

import ipywidgets
```

```python
import umap
```

```python
amd33_raw = np.load("./acam33_rxn_fps.npy")
```

```python
# %%time
all_n_n = [10,50,100,200,500,1000]
all_n_n = [10]

for n_n in all_n_n:
    
    reducer = umap.UMAP(n_neighbors=n_n, init='random',low_memory=False)
    embedding = reducer.fit_transform(amd33_raw)
    np.save(f"umap_acid_amine_33_nn_{n_n}.npy",embedding)
    reducer = []
    embedding = []
```

```python
# %%time
# all_n_n = [10,50,100,200,500]

for n_n in all_n_n:
    
    # this will look different because umap runs differently for every iteration
    embedding = np.load(f"umap_acid_amine_33_nn_{n_n}.npy")
    
    plt.figure(figsize=(10,8))
    plt.scatter(embedding[:,0],embedding[:,1])
    plt.title(str(n_n))
```
