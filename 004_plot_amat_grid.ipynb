```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.axes_grid1 import make_axes_locatable
import pprint
from matplotlib import cm
from matplotlib.colors import ListedColormap
```

```python
def blank_amat(alength):
    """
    produces blank adjacency matrix of all zeros
    with length = alength
    """
    matt = []
    for i in range(alength):
        matt.append([0]*alength)
    return matt
```

```python
def amat_entry(amat,r,c,val):
    """
    changes the (r,c) and (c,r) values of matrix "amat" to "val"
    """
    amat[r-1][c-1] = amat[c-1][r-1]= val
    return
```

```python
# make the plot

def plot_save_amat(amat,have_cbar=False,save_name=None):

    amat_len = len(amat)
    x = list(range(1,(amat_len+1)))
    y = list(range(1,(amat_len+1)))

    fig, ax = plt.subplots(figsize=(2,2),dpi=900)
    im = ax.imshow(amat,cmap=use_cmap)

    ax.set_xticks(np.arange(len(x)))
    ax.set_yticks(np.arange(len(y)))

    ax.set_xticklabels(x,fontname='Arial',fontsize=ax_fontsize)
    ax.set_yticklabels(y,fontname='Arial',fontsize=ax_fontsize)
    im.set_clim(vmin, vmax)
    
    # Minor ticks
    ax.set_xticks(np.arange(-.5, amat_len, 1), minor=True);
    ax.set_yticks(np.arange(-.5, amat_len, 1), minor=True);

    gg = 0.8
    # Gridlines based on minor ticks
    ax.grid(which='minor', color=[gg,gg,gg,1], linestyle='-', linewidth=1)
    

    
    if have_cbar:
        divider = make_axes_locatable(ax)
        cax = divider.append_axes("right", size="5%", pad=0.05)
        cb = plt.colorbar(im, cax=cax)
        cb.ax.tick_params(labelsize=18)
        cb.set_ticks([-2,-1,0,1,2,3])
        cb.ax.set_yticklabels([-2,-1,0,1,2,3],font="Arial",fontsize=cbar_fontsize)
        



    plt.subplots_adjust(wspace=0, hspace=0)
    plt.tight_layout()
    if save_name: plt.savefig(save_name,dpi=900)
    pass
    # 
```

```python
plot_save_amat(ac3_am3,True,"./figures/fig2a_new_1.png")
plot_save_amat(ac3_am3,True,"./figures/fig2c_new_2.png")
```
